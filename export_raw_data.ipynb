```python
import pandas as pd

def export_raw_data(excel_file):
    # 假设你的 Sheet 顺序是 [年, 专项, 周, 打球记录, 装备记录]
    # 或者通过名字读取更为保险
    try:
        # 1. 提取打球开销记录 (Sheet 4)
        sessions_df = pd.read_excel(excel_file, sheet_name=4) # 索引从0开始，3是第4个
        sessions_df.to_csv("sessions.csv", index=False, encoding='utf-8-sig')
        
        # 2. 提取装备开销记录 (Sheet 5)
        equipment_df = pd.read_excel(excel_file, sheet_name=3) # 4是第5个
        equipment_df.to_csv("equipment.csv", index=False, encoding='utf-8-sig')
        
        print("✅ 转换完成！原始数据库 sessions.csv 和 equipment.csv 已就绪。")
    except Exception as e:
        print(f"❌ 转换失败: {e}。请检查 Sheet 顺序或名称。")
```

```python
# 执行转换
export_raw_data("羽毛球开销记录.xlsx")
```
